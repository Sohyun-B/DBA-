## **행과 열로 접근하여 데이터 추출하기**

```python
# 열을 통한 접근
tt['survived'] #1차원 배열
tt[['survived', 'pclass']] #2차원 배열
```

```python
# 행을 통한 접근
# loc: 행 인덱스 기준으로 추출
#     -1로 끝값 접근 불가
# iloc: 추출한 인덱스 기준
#     -1로 끝값 접근 가능


# 5행의 모든 데이터값 출력
tt.loc[5]
tt.iloc[5]
```

```python
num_row=tt.shape[0] #행의 수
last_index=num_row-1 #행의 수 -1 => 가장 마지막 행
tt.loc[last_index] #가장 마지막 데이터 추출
```

```python
#원하는 행만 숫자로 추출 가능
tt.loc[[0, 2, 3,6, 880]]
tt.iloc[[0, 2, 3,6, 880]] #데이터를 자르지 않았으므로 iloc도 값 동일
```

```python
#10부터 150까지 데이터를 2 간격으로
tt.loc[10:150:2]
tt.iloc[10:150:2]
```

```python
# loc과 iloc의 차이 중 하나: 컬럼 접근
# iloc: 컬럼을 숫자로만 접근 가능
# loc: 컬럼 이름으로 접근 가능
# ex) tt.loc[행, 열]

tt.loc[::2, ['age', 'survived']] #::2가 행 관련, ['age']가 열관련
```

# **기초 통계값 사용**

```python
# 기초 통계

# 평균
tt['fare'].mean() #판다스 사용
np.mean(tt['fare']) #넘파이 사용

# 합
tt['fare'].sum()
np.sum(tt['fare'])

# 중간값
tt['fare'].median()
np.median(tt['fare'])

# 최댓값
tt['fare'].argmax()
np.argmax(tt['fare'])

# 최솟값
tt['fare'].argmin()
np.argmin(tt['fare'])
```

```python
# groupby('그룹을 묶을 컬럼')['통계치를 볼 컬럼'] 그룹별로 묶음
tt.groupby('pclass')['fare', 'age'].mean()
```

```python
# 접근 순서

# 열 -> 행
tt[['survived', 'class']].loc[2: 40]

# 행 -> 열
tt.loc[2: 40][['survived', 'class']]
```

```python
# query('요구조건') 데이터 추출 조건
# 논리 연산자 and or 사용 가능


# embark가 S인 데이터 접근
tt.query('''embarked == "S" ''')

# 수치형 데이터 fare 접근
tt.query("fare > 32")

#S가 아니고 요금 90 초과 승객이나 나이가 10살 미만인 승객의 생존률의 평균
tt.query(''' embarked != "S" and fare > 90 or age < 10''')['survived'].mean()
```
