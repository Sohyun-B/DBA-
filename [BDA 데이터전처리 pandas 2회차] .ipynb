```python
#필요한 데이터셋 다운로드
import seaborn as sns
import pandas as pd
tt=sns.load_dataset('titanic')
```

```python
#나이를 기준으로 정렬
#디폴트: 오름차순
#ascending을 사용하여 내림차순으로 변경
tt.sort_values('age', ascending=False)
#tt.sort_values('age', descending=True) 이건 불가능
```

**파생변수 만들기**

```python
tt['sex_oneshot']=0 #새로운 컬럼 생성 후 0으로 초기화
#남성이면 0, 여성이면 1로
for i in range(len(tt)):
    if tt.sex.loc[i]=="male":
        tt['sex_oneshot'].loc[i]=0
    elif tt.sex.loc[i]=="female":
        tt['sex_oneshot'].loc[i]=1
```

```python
print(tt.sex_oneshot.value_counts())
print("==================")
print(tt.sex.value_counts())
```

```python
tt['fare_100']=tt['fare']+100  #연산 후 데이터를 바로 넣어서 파생변수를 만들 수도 있
```

```python
import numpy as np #넘파이의 np.where 함수 사용
```

**파생변수 관련 함수**
*   **np.where(조건)** #조건에 맞는 인덱스를 호출
*   **dataframe.assign(조건, '조건충족시', '조건미충족시')** #조건에 따라 값을 할당



```python
#np.where는 조건에 맞는 인덱스를 호출하는 함수
np.where(tt['fare']>=100)
```

```python
#np.where(tt['fare']>=10이 조건
#조건을 충족하면 H, 아니면 L를
#fare_cls라는 행을 만들어서 저장
tt.assign(fare_cls=np.where(tt['fare']>=10, 'H', 'L'))
```

```python
tt.assign(fare_100 = lambda x : x['fare_100']/ x['fare'])
```

*   **df.groupby(컬럼)**
*   **그룹이름.get_group(원하는 그룹)**
*   **df.groupby('그룹할컬럼1').agg(변수명 = ('기준컬럼1','기준컬럼을 볼 통계치'), ...)**



```python
grouped=tt.groupby('class')

#데이터값 확인
for i, j in grouped:
    print(i)
    print(j.head(10))
```

```python
grouped.get_group('First') #'class'로 묶인 것 중에 First인 것들만 호출
```

```python
groups=tt.groupby(['class','embarked'])
for i, j in groups:
    print(i)
    print(j.head(10))
```

```python
groups.mean() #묶인 것 별로 평균을 구함
```

```python
groups.std()
```

```python
tt.groupby('class').agg(fare_sum=('fare', 'sum'), #class를 기준으로 그룹하여 fare의 sum을 구함
                        fare_mean=('fare', 'mean'),
                        fare_median=('fare', 'median'),
                        n_cnt=('fare', 'count'))
```

```python
#1개의 컬럼에 여러 개의 통계치를 볼 때.
grouped.agg({'fare':['sum', 'mean', 'median', 'count']})
```

```python
def min_max(x):
  return x.max()-x.min()
```

```python
grouped.agg(min_max)
```

## **데이터 병합**
*  **df.concat(데이터1, 데이터2, how='left', on='컬럼명'): **

    -덩어리로 합침

    -행으로 붙이는 것/열로 붙이는 것

*   **df.merge([데이터1, 데이터2, 데이터3,...], axis=0):**

    -join (left, right, outer)



```python
#예시 데이터프레임 생성

test1=pd.DataFrame({'학회원번호': [1,2,3,4,5],
                    '학회원분반': ['python','SQL','R','데이터분석입문','데이터분석전처'],
                    })

test2= pd.DataFrame({'학회원번호':[1,2,3,4,5],
                    '벌점현황':[5,5,10,0,3]})
```

```python
pd.merge(test1, test2, how='inner', on='학회원번호')
pd.merge(test1, test2, how='outer', on='학회원번호')
pd.merge(test1, test2, how='left', on='학회원번호')
```

```python
pd.concat([test1, test2], axis=0) #데이터프레임이 그대로 세로로 붙음
```

```python
pd.concat([test1, test2, test1, test1], axis=1) #데이터프레임이 그대로 가로로 붙음
```
